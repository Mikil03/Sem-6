```python
median_val = df['Arrival Delay in Minutes'].median()
# Filling missing values with median as distribution of arrival delay was heavily skewed
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(median_val)
```

```python
df2 = pd.get_dummies(df, columns = ['Gender','Customer Type','Class','Type of Travel'])
print(df2)
```

```python
df2.info()
```

```python
from sklearn.model_selection import train_test_split
X = df2.drop(columns = ['id','satisfaction'])
Y = df2['satisfaction']
train_X, test_X, train_y, test_y = train_test_split(X, Y, test_size=0.35, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_X,train_y)

#predicting test data
pred_y = model.predict(test_X)

#accuracy score
from sklearn.metrics import accuracy_score
score_lr = accuracy_score(test_y,pred_y)
print(score_lr*100,"%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
knn = KNeighborsClassifier(p=2, metric='minkowski', n_neighbors=5)
knn.fit(train_X,train_y)
y_knn_pred = knn.predict(test_X)
print(y_knn_pred)

cm_knn = confusion_matrix(test_y, y_knn_pred)
print (cm_knn)
score_knn = accuracy_score(test_y, y_knn_pred)
print (score_knn)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(train_X,train_y)

#predicting test data
pred_y = rfc.predict(test_X)
#accuracy score
from sklearn.metrics import accuracy_score
score_rf = accuracy_score(test_y,pred_y)
score_rf
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dt.fit(train_X, train_y)

y_pred = dt.predict(test_X)


score_dt = accuracy_score(test_y, y_pred)
print (score_dt)
```

```python
model_names = ['Decision Tree','Logistic Regression', 'KNN', 'Random Forest' ]
accuracy_scores = [score_dt, score_lr, score_knn, score_rf] 
df = pd.DataFrame({'Model': model_names, 'Accuracy': accuracy_scores})
sns.barplot(x='Model', y='Accuracy', data=df)
plt.xticks(rotation=90)
#show values on top
for index, row in df.iterrows():
    plt.text(row.name,row.Accuracy, round(row.Accuracy,3), ha="center")
plt.show()
```
