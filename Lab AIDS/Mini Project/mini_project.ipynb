```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv", index_col=0)
df.head()
```

#### Data Visualization

```python
cols = ['Gender', 'Customer Type', 'Type of Travel', 'Class']
fig, ax = plt.subplots(2,2, figsize=(12,10))

sns.countplot(data=df, x=cols[0], ax = ax[0][0])
sns.countplot(data=df, x=cols[1], ax = ax[0][1])
sns.countplot(data=df, x=cols[2], ax = ax[1][0])
sns.countplot(data=df, x=cols[3], ax = ax[1][1])
plt.show()
```

```python
sns.countplot(x=df.satisfaction)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True))
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
sns.scatterplot(df, x= 'Departure Delay in Minutes', y='Arrival Delay in Minutes')
plt.show()
```

```python
d1 = df.pivot_table(index='Type of Travel',columns= 'Class', values='Age', aggfunc='mean')
sns.heatmap(data=d1)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
