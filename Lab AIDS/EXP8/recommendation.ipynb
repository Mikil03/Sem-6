```python
df_anime['episodes'] = df_anime['episodes'].replace('Unknown',None)
```

```python
df_anime['episodes'] = pd.to_numeric(df_anime['episodes'])
```

```python
df_anime.sort_values(by=['members'],ascending=False, inplace=True)
```

```python
df_anime.reset_index(inplace=True, drop=True)
```

```python
df_anime = df_anime[:100]
```

```python
df_anime
```

```python
index = df_anime.anime_id
```

```python
ratings = pd.read_csv("rating.csv")
ratings
```

```python
ratings.drop(ratings[ratings['rating']<1].index, inplace=True)
```

```python
df_ratings = pd.DataFrame(columns=ratings.columns)
df_ratings = df_ratings.append({'user_id':-1, 'anime_id':-1, 'rating':-1}, ignore_index=True)
```

```python
df_ratings
```

```python
for i in index:
    df_ratings = pd.concat([df_ratings, ratings[ratings['anime_id']==i]], axis=0)
```

```python
df_ratings.head()
```

```python
df_ratings.drop([0], inplace=True)
```

```python
df_ratings.head()
```

```python
df_ratings.sort_values(by=['user_id'],ascending=False, inplace=True)
```

```python
df_ratings_new = df_ratings[:100000]
df_ratings_new.anime_id.nunique()
```

```python
pivot = df_ratings_new.pivot_table(index='anime_id', columns='user_id', values='rating')
```

```python
pivot = pivot.fillna(0)
pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pivot)
similarity_scores.shape
```

```python
###

def recommend(anime_name):
    # index fetch
    anime = df_anime.anime_id[df_anime['name'] == anime_name].values[0]
    index = np.where(pivot.index==anime)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = df_anime.name[df_anime['anime_id'] == pivot.index[i[0]]].values[0]
        data.append(item)
    
    return data
```

```python
recommend('Fullmetal Alchemist: Brotherhood')
```
